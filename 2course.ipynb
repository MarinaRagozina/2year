##  Обработка файла с шаблонами

```
def read_infile(infile):
    patterns_dict=[]#(pattern,case)
    with open(infile, "r", encoding="utf8") as fin:
        for line in fin:
          line = line.strip()
          line, other = line.split('\t')
          case, pattern_corp = line.split(' ', maxsplit=1)#сплит по первому пробелу
          case = re.sub('и', 'nomn', case)
          case = re.sub('р', 'gent', case)
          case = re.sub('д', 'datv', case)
          case = re.sub('в', 'accs', case)
          case = re.sub('т', 'ablt', case)
          case = re.sub('п', 'loct', case)
          pattern=(pattern_corp,case)
          patterns_dict.append(pattern)
    return patterns_dict
```

```
numerals=['два', 'три','четыре','оба','полтора']
```

```
def inflection(word,case):
  parse = analyzer.parse(word) [0]
  new_word = parse.inflect({case}).word
  return new_word
```

```
def accs_inflection(word,number,animacy,case):
  parse = analyzer.parse(word) [0]
  if number == 'sing' and case == 'accs':
    if animacy == 'anim':
      new_word = parse.inflect({'gent','sing'}).word
    else:
      new_word = parse.inflect({'nomn','sing'}).word
  else:
    new_word = parse.inflect({case, number}).word
  return new_word
      
```

```
def accs_num_inflection(word,animacy,case):
  parse = analyzer.parse(word) [0]
  if case == 'accs':
    if animacy == 'anim':
      new_word = parse.inflect({'gent','NUMR'}).word
    else:
      new_word = parse.inflect({'nomn','NUMR'}).word
  else:
    new_word = parse.inflect({case, 'NUMR'}).word
  return new_word
```

```
def case_change(string,parent,processed_case):#изменение формы
  new_word = string[1].lower()
  if 'Case' in string[5]:
    if 'Animacy' in string[5]: #при наличии одушевленности в грамматических метках
      number = string[5]['Number']
      gender = string[5]['Gender']
      animacy = string[5]['Animacy']
      parse = analyzer.parse(new_word) [0]
      if gender!='femn':
        new_word = accs_inflection(new_word,number,animacy,processed_case)
      else:
        new_word = inflection(new_word,processed_case)
    elif 'Animacy' in parent[5]:  #при наличии одушевленности в грамматических метках root  
      if string[3] == 'NUM' and string[2]!='один':
        if string[2] in numerals:#малые числительные
          new_word = accs_num_inflection(new_word, parent[5]['Animacy'], processed_case)
        elif 'Coll' in analyzer.parse(new_word)[1].tag and processed_case=='accs': #собирательное числительное для одушевленности с существительным (трое ребят)
          new_word = analyzer.parse(new_word)[1].inflect({'accs','anim'}).word
        else:
          new_word = inflection(new_word,processed_case)
      else:
        if 'Number' in string[5]:
          number = string[5]['Number']
        if 'Animacy' in string[5]:
          animacy = parent[5]['Animacy']
        if string[5].get('Gender') != 'femn' and 'Number' in string[5] and 'Animacy' in string[5]:#так как gender может отсутствовать
          new_word = accs_inflection(new_word,number,animacy,processed_case)
        else:
          new_word = inflection(new_word,processed_case)
    elif string[3] == 'NUM' and processed_case == 'accs' and string[7] in ['root','conj'] and 'Coll' in analyzer.parse(new_word)[1].tag:
      #для собирательных числительных в UDPipe нет метки, поэтому нет учета одушевленности: здесь учитываем для примеров типа "трое" без существительного
      new_word = analyzer.parse(new_word)[1].inflect({'accs','anim'}).word
    else:
      new_word = inflection(new_word,processed_case)
  return new_word
```

```
def num_inflection(word, nums, case):
  cur_governer = []
  for elem in nums:# поиск числительного, которое управляет root
    if elem[6] == word[0] or (elem[6]==word[6]):
      if elem[7] == 'conj':
        #if word[5]['Case'] == parent[5]['Case']:# проверяем, относится ли данный conj к числительному
        if cur_governer == [] or cur_governer[0] < elem[0]:
          cur_governer = elem
      elif cur_governer == [] or cur_governer[0] < elem[0]:
          cur_governer = elem
  if cur_governer != []:
    if case == 'nomn' or case == 'accs':
      if cur_governer[2] in numerals and word[7]!='amod' and word[7]!='acl':# если управляющее числительное малое
        new_word = word[1]
        new_word = inflection(new_word,'gent')
        #if case == 'nomn' or (case == 'accs' and (parent[5]['Animacy'] == 'Inan' or 'Animacy' not in parent[5])):
        if case == 'nomn' or (case == 'accs' and (word[5]['Animacy'] == 'Inan' or 'Animacy' not in word[5])):
          new_word = inflection(new_word,'sing')
        else:
          new_word = inflection(new_word,'plur')
      else:
        new_word = inflection(word[1],'gent')
    else:
      new_word = inflection(word[1],'plur')
      new_word = inflection(new_word, case)
  return new_word
```

```
new_strings=[]
for string in strings:
  string_change = []#список слов + необходимые изменения
  root_number = int(string[0][0])
  for elem in string[1]:
    cur_number = int(elem[0])
    if elem[3] == 'PRON' and elem[7] == 'nmod': #случаи типа 'ее'
      string_change.append([elem[1],'None']) #вид для неизменямых слов
      
    elif ((cur_number < root_number or elem[7] == ("appos") or (elem[7] == ("acl")) or (elem[7] == ("conj"))) 
          and int(elem[6])== root_number) or cur_number == root_number: # поиск root, модификаторов слева, приложений и однородных членов
      if string[3] != [] and elem[3] != 'NUM':
        if elem[5]['Case'] == 'nom' and elem[7] == "conj":
          string_change.append([elem,elem,'case_change']) #вид для именных групп с числительным и несогласованными с ним однородными членами "три девочки и мальчик"
        else:
          string_change.append([elem,string[3],'num_inflection']) #вид для именных групп с числительными
      else:
        string_change.append([elem,string[0],'case_change']) #вид для именных групп
        
    elif string[2] != [] and elem[3] != 'PUNCT' and elem[3] != 'CCONJ': #поиск зависимых для приложений и однородных членов
      current = []
      for each in string[2]:
        if int(elem[6]) == int(each[0]):
          current.append([elem,each,'case_change'])
      if current != []:
        string_change.append([elem,each,'case_change'])
      else:
        if elem[3]!='PROPN':
          elem[1] = elem[1].lower()
        string_change.append([elem[1],'None'])
        
    else: #неизменяемые части
      if elem[1] != '.': #удаляем точки
        if elem[3]!='PROPN':
          elem[1] = elem[1].lower()
        string_change.append([elem[1],'None']) #вид для неизменямых слов
  new_strings.append(string_change)
for string in new_strings:
  print(string)
```

```
full_patterns = {}
for pattern in patterns_dict:
  #pattern_ud = full_ud_model.process(pattern[0])
  #print(pattern_ud)
  final_lines = []
  processed_case = pattern[1] #данный падеж
  order_sentences = []
  for string in new_strings:
    final_line = ''
    for elem in string:
      if elem[-1] == 'None':#неизменяемое слово
        cur_word = elem[0]
      if elem[-1] == 'case_change': #изменяемое слово
        cur_word = case_change(elem[0],elem[1],processed_case)
        if elem[0][3]=='PROPN':
          cur_word=cur_word.title()
      if elem[-1] == 'num_inflection':# для зависимых от числительного слов
        cur_word = num_inflection(elem[0],elem[1],processed_case)
        if elem[0][3] == 'PROPN':
          cur_word = cur_word.title()
      if final_line == "" or cur_word == ',':
        final_line += cur_word
      else:
        final_line+=' '+cur_word
    final_lines.append(final_line)
  full_patterns[pattern[0]]=final_lines
  
with open(os.path.join(SOURCE_DIR, "templates_rus_case_output.txt"), 'w', encoding='utf8') as fout:
  for key, value in full_patterns.items():
    for elem in value:
      if re.search(r'XXX', key):
        result = re.sub(r'XXX',elem,key)
      else:
        result = re.sub(r'ХХХ',elem,key)#ХХХ - кириллицей  
      fout.write('{}\n'.format(result))
```
